```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
```

```python
df= pd.read_csv('Heart Prediction Quantum Dataset.csv')
df.head()
```

```python
for col in df.columns:
    print(f'{col}: {df[col].nunique()}')
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='HeartDisease', data=df, hue='HeartDisease', palette='viridis', legend=False)
plt.title('Distribution of Heart Disease')
plt.show()
```

```python
categorical_cols = ['Gender']

for col in categorical_cols:
    plt.figure(figsize=(12, 6))
    sns.countplot(x=col, data=df, palette='viridis', hue='HeartDisease')
    plt.title(f'Heart Disease by {col}')
    plt.show()
```

```python
numerical_cols = ['Age', 'BloodPressure', 'Cholesterol', 'HeartRate', 'QuantumPatternFeature']

for col in numerical_cols:
    plt.figure(figsize=(12, 6))
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
plt.figure(figsize=(14, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
for col in numerical_cols:
    plt.figure(figsize=(12, 6))
    sns.boxplot(x='HeartDisease', y=col, data=df, palette='viridis',legend=False)
    plt.title(f'{col} vs. Heart Disease')
    plt.show()
```
